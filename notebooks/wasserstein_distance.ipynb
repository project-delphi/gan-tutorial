

## Wasserstein Distance

The Wasserstein distance is another metric function that is used to measure the difference between two probability distributions. The Wasserstein distance is defined as:

$W(P, Q) = \inf_{\Pi(P, Q)} \int c(x, y) d\Pi(x, y)$

where $\Pi(P, Q)$ is the set of all joint probability distributions 
with marginals $P$, $Q$ and:

* $c(x, y)$ is the cost of transporting a unit of mass from $x$ to $y$.

### Wasserstein Loss Formula for GANs

In the context of Generative Adversarial Networks (GANs), the Wasserstein distance is used as a loss function to measure the difference between the generated and real data distributions. The Wasserstein loss function is defined as:

$W(P_r, P_g) = \inf_{\gamma \in \Pi(P_r, P_g)} \mathbb{E}_{(x, y) \sim \gamma} [||x - y||]$

where $P_r$ is the real data distribution, $P_g$ is the generated data distribution, and $\Pi(P_r, P_g)$ is the set of all joint probability distributions with marginals $P_r$ and $P_g$.

```python
from scipy.stats import wasserstein_distance

# Example distributions
distribution1 = [0.0, 1.0, 2.0]
distribution2 = [2.0, 2.0, 2.0]

# Calculate Wasserstein distance
distance = wasserstein_distance(distribution1, distribution2)
print(f'Wasserstein Distance: {distance}')
import matplotlib.pyplot as plt

# Plot the distributions
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.hist(distribution1, bins=3, alpha=0.5, label='Distribution 1')
plt.hist(distribution2, bins=3, alpha=0.5, label='Distribution 2')
plt.legend(loc='upper right')
plt.title('Histograms of Distributions')

plt.subplot(1, 2, 2)
plt.plot(distribution1, label='Distribution 1', marker='o')
plt.plot(distribution2, label='Distribution 2', marker='o')
plt.legend(loc='upper right')
plt.title('Line Plots of Distributions')

plt.suptitle(f'Wasserstein Distance: {distance}')
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import seaborn as sns

# Convert lists to numpy arrays for easier manipulation
distribution1_np = np.array(distribution1)
distribution2_np = np.array(distribution2)

# Plot the CDFs
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
sns.ecdfplot(distribution1_np, label='Distribution 1')
sns.ecdfplot(distribution2_np, label='Distribution 2')
plt.legend(loc='lower right')
plt.title('CDF of Distributions')

# Plot the PDFs
plt.subplot(1, 2, 2)
sns.kdeplot(distribution1_np, label='Distribution 1', fill=True)
sns.kdeplot(distribution2_np, label='Distribution 2', fill=True)
plt.legend(loc='upper right')
plt.title('PDF of Distributions')

plt.suptitle(f'Wasserstein Distance: {distance}')
plt.tight_layout()
plt.show()
```

```python
import numpy as np

def wasserstein_distance(P, Q):
    """
    Compute the Wasserstein distance (1D) between two distributions P and Q.

    Parameters:
    P (ndarray): 1D array of probabilities or histogram bins for distribution P.
    Q (ndarray): 1D array of probabilities or histogram bins for distribution Q.

    Returns:
    float: The Wasserstein distance between P and Q.
    """
    # Normalize the histograms (make sure they sum to 1)
    P = P / np.sum(P)
    Q = Q / np.sum(Q)
    
    # Compute cumulative distribution functions (CDF)
    cdf_P = np.cumsum(P)
    cdf_Q = np.cumsum(Q)
    
    # Compute Wasserstein distance (L1 distance between the CDFs)
    return np.sum(np.abs(cdf_P - cdf_Q))

# Example distributions (histograms)
P = np.array([0.1, 0.3, 0.4, 0.2])
Q = np.array([0.2, 0.1, 0.3, 0.4])

# Compute Wasserstein distance
distance = wasserstein_distance(P, Q)
print("Wasserstein distance:", distance)
```

```python
from scipy.stats import wasserstein_distance
import numpy as np

# Example 3D distributions
distribution1_3d = np.random.rand(100, 3)
distribution2_3d = np.random.rand(100, 3)

# Calculate Wasserstein distance for each dimension
distances = [wasserstein_distance(distribution1_3d[:, i], distribution2_3d[:, i]) for i in range(3)]
print(f'Wasserstein Distances for each dimension: {distances}')

# Calculate the average Wasserstein distance across all dimensions
average_distance = np.mean(distances)
print(f'Average Wasserstein Distance: {average_distance}')
```
