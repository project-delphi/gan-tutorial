# Some Fixed Point Theory

## Introduction

In mathematics, a fixed point (sometimes shortened to fixpoint, also known as an invariant point) of a function is an element of the function's domain that is mapped to itself by the function. That is to say, a point x is a fixed point of a function f if f(x) = x.


```python
import sympy as sp

# Define the variable and the function
x = sp.symbols('x')
f = x**2 - 2*x + 2

# Solve for the fixed point
fixed_points = sp.solve(f - x, x)
print(f'Function f is : {f}')
print(f'Solving for fixed points: {fixed_points}')
print(f'verification of fixed points')
for point in fixed_points:
    print(f'f({point}) = {f.subs(x, point)}')
```

```python
import sympy as sp

# Define the variables
x, y = sp.symbols('x y')

# Define the first set of functions
f1 = x**2 - 2*x*y + y**2 + 2
f2 = x*y - x + y**2 - 1

# Solve for the fixed points of the first set of functions
fixed_points_2d = sp.solve([f1 - x, f2 - y], (x, y))

# Display results for the first set of functions
print(f'Functions are: f1 = {f1}, f2 = {f2}')
print(f'Solving for fixed points yields {len(fixed_points_2d)} fixed points:')
for fixed_point in fixed_points_2d:
    print(f'({fixed_point[0]}, {fixed_point[1]})')
print('Verification of fixed points for f1 and f2:')
for point in fixed_points_2d:
    f1_val = f1.subs({x: point[0], y: point[1]})
    f2_val = f2.subs({x: point[0], y: point[1]})
    print(f'f1({point[0]}, {point[1]}) = {f1_val}')
    print(f'f2({point[0]}, {point[1]}) = {f2_val}')
```

```python
# Define the second set of functions
h1 = x**3 - 3*x*y + y**3
h2 = x**2 - y**2 + x*y

# Solve for the fixed points of the second set of functions
fixed_points_h = sp.solve([h1 - x, h2 - y], (x, y))

# Display results for the second set of functions
print(f'\nFunctions are: h1 = {h1}, h2 = {h2}')
print(f'Solving for fixed points: {fixed_points_h}')
print('Verification of fixed points for h1 and h2:')
for point in fixed_points_h:
    h1_val = h1.subs({x: point[0], y: point[1]})
    h2_val = h2.subs({x: point[0], y: point[1]})
    print(f'h1({point[0]}, {point[1]}) = {h1_val}')
    print(f'h2({point[0]}, {point[1]}) = {h2_val}')
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the logistic map function
def logistic_map(x, r):
    return r * x * (1 - x)

# Find the fixed point
def find_fixed_point(r):
    return (r - 1) / r  # For the logistic map: f(x) = x => x = (r-1)/r

# Generate the sequence of iterates
def iterate_logistic_map(x0, r, num_iterations=100):
    x_values = [x0]
    for _ in range(num_iterations):
        x0 = logistic_map(x0, r)
        x_values.append(x0)
    return x_values

# Parameters
r = 3.5  # Chosen value for r (this can be varied)
x0 = 0.1  # Initial value for x
num_iterations = 100

# Find the fixed point for the given r
fixed_point = find_fixed_point(r)

# Iterate the logistic map
x_values = iterate_logistic_map(x0, r, num_iterations)

# Plot the results
plt.figure(figsize=(10, 6))
plt.plot(x_values, label='Logistic Map Iterates', color='b')
plt.axhline(fixed_point, color='r', linestyle='--', label=f'Fixed Point x* = {fixed_point:.2f}')
plt.title(f'Logistic Map Iterations (r = {r})')
plt.xlabel('Iterations')
plt.ylabel('x')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import sympy as sp

# Define the variables
x, y = sp.symbols('x y')

# Define the functions
f1 = x - y**2
f2 = y - x**2

# Solve for the fixed points
fixed_points = sp.solve([f1 - x, f2 - y], (x, y))

# Display the fixed points
print(f"Fixed points: {fixed_points}")

# Function to iterate over the system
def iterate_system(x0, y0, num_iterations=100):
    x_values, y_values, z_values = [x0], [y0], [0]  # Initialize with iteration 0
    
    for i in range(1, num_iterations + 1):
        x_new = x_values[-1] - y_values[-1]**2  # f1(x, y) = x - y^2
        y_new = y_values[-1] - x_values[-1]**2  # f2(x, y) = y - x^2
        
        x_values.append(x_new)
        y_values.append(y_new)
        z_values.append(i)  # Track the iteration step as z
    
    return x_values, y_values, z_values

# Initial guess
x0, y0 = 0.5, 0.5

# Iterate the system
num_iterations = 100
x_vals, y_vals, z_vals = iterate_system(x0, y0, num_iterations)

# Plotting the 3D trajectory
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Plot the iteration path in 3D
ax.plot(x_vals, y_vals, z_vals, label="Iteration Path", color='blue', marker='o')

# Plot the fixed points
for point in fixed_points:
    ax.scatter(point[0], point[1], 0, color='red', s=100, label=f'Fixed Point: {point}')

# Set plot details
ax.set_title("Fixed Point Iteration in 2D System (3D View)")
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("Iteration Step")
ax.legend(loc="best")
ax.grid(True)

# Show plot
plt.show()
```

## Brouwer's Fixed-Point Theorem

Brouwer's Fixed-Point Theorem states that:

> *Every continuous function from a compact convex set to itself has at least one fixed point.*

Formally, let $X$ be a **non-empty compact convex set** in a Euclidean space $\mathbb{R}^n$. If $f: X \to X$ is a continuous function, then there exists at least one point $x^* \in X$ such that:

$$
f(x^*) = x^*
$$

#### Example:

For $X = [0, 1] \subset \mathbb{R}$ and a continuous function $f: [0, 1] \to [0, 1]$, Brouwer's theorem guarantees the existence of some $x^* \in [0, 1]$ where:

$$
f(x^*) = x^*
$$
