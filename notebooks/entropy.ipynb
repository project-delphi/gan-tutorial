# Entropy

## Introduction

Entropy is a measure of the uncertainty associated with a random variable. In the context of machine learning, entropy is commonly used as a measure of the uncertainty in a dataset. The entropy of a dataset is zero when the dataset contains instances of only one class. The entropy of a dataset is high when the dataset contains instances of many different classes.


## Entropy of a dataset
The entropy of a dataset is calculated as follows:

$$ H(S) = - \sum_{i=1}^{c} p_i \log_2(p_i) $$

where $H(S)$ is the entropy of the dataset, $S$, $c$ is the number of classes in the dataset, and $p_i$ is the proportion of instances in class $i$ in the dataset.

## Why the log?

The log function is used in the entropy formula because it is a common choice for the function that measures the uncertainty associated with a random variable. The log function is used because it is a monotonic function that increases as the probability of an event decreases. This makes it a good choice for measuring the uncertainty associated with a random variable. For example, if the probability of an event is 1, the log of the probability is 0, which means that there is no uncertainty associated with the event. If the probability of an event is close to 0, the log of the probability is a large negative number, which means that there is a lot of uncertainty associated with the event.

## Why the log2?

The log base 2 is used in the entropy formula because it is a common choice for the base of the log function in information theory. The log base 2 is used because it allows the entropy to be measured in bits. This makes it easier to interpret the entropy as a measure of the uncertainty in the dataset. For example, if the entropy of a dataset is 1 bit, it means that there is one bit of uncertainty associated with each instance in the dataset.

## Why the negative sign?

The negative sign in the entropy formula is used to ensure that the entropy is always positive. The negative sign is used because the log function is always negative for probabilities less than 1. By multiplying the log function by -1, the entropy is always positive, which makes it easier to interpret as a measure of the uncertainty in the dataset.

## Why weight by the probability?

The entropy formula weights the log of the probabilities by the probabilities themselves. This is done to ensure that the entropy captures the uncertainty in the dataset. If the log of the probabilities were not weighted by the probabilities themselves, the entropy would not be able to capture the uncertainty in the dataset. For example, if the log of the probabilities were not weighted by the probabilities themselves, the entropy of a dataset with two classes would always be 1, regardless of the probabilities of the classes.





```python
import numpy as np

import matplotlib.pyplot as plt

# Define the function -p * log2(p)
def plogp(p):
    return -p * np.log2(p)

# Generate p values from 0 to 1
p_values = np.linspace(0, 1, 100)
plogp_values = [plogp(p) if p > 0 else 0 for p in p_values]  # Handle log(0) case

# Plot the function
plt.plot(p_values, plogp_values)
plt.xlabel('p')
plt.ylabel('-p log2(p)')
plt.title('-p log2(p) as a function of p')
plt.grid(True)
plt.show()
```

```python
# Example 3: Varied entropy for classification of k-mers
kmer_counts = {'AAA': 50, 'AAT': 30, 'ATA': 20, 'ATT': 10}
total_count = sum(kmer_counts.values())
p3 = [count / total_count for count in kmer_counts.values()]
entropy_varied = -np.sum([p * np.log2(p) for p in p3 if p > 0])
print(f"Varied entropy for classification of k-mers: {entropy_varied}")
```

```python
def binary_entropy(p):
    if p == 0 or p == 1:
        return 0
    return -p * np.log2(p) - (1 - p) * np.log2(1 - p)

# Example usage:
p = 0.5  # probability of one class
entropy = binary_entropy(p)
print(f"Entropy for p = {p}: {entropy}")
import matplotlib.pyplot as plt

# Plot entropy as a function of p for binary classification
p_values = np.linspace(0, 1, 100)
entropy_values = [binary_entropy(p) for p in p_values]

plt.plot(p_values, entropy_values)
plt.xlabel('p')
plt.ylabel('Entropy')
plt.title('Entropy as a function of p for binary classification')
plt.grid(True)
plt.show()
```

## Variance and Entropy Calculations for Normally Distributed Data

In this section, we will calculate the variance and entropy for a normally distributed dataset. We will also visualize the data using density plots.


For a normal distribution with mean and variance $\mu$ and $\sigma^2$ vectors, the entropy is given by:

$$ H(X) = \frac{1}{2} \log(2\pi e \sigma^2) $$

where $H(X)$ is the entropy of the normal distribution, $\pi$ is the mathematical constant pi, $e$ is the mathematical constant e, and $\sigma^2$ is the variance of the normal distribution.

### Generating Normally Distributed Data

We will generate a dataset of 1000 samples from a normal distribution with a mean of 0 and a standard deviation of 1.



In the code above, we first generate the normally distributed data using `np.random.normal()`. We then calculate the variance using `np.var()`. For entropy, we use the `scipy.stats.norm` module to fit a normal distribution to the data and calculate its entropy. Finally, we plot the density of the data using `seaborn.histplot()` with the `kde=True` option to include the kernel density estimate.


```python
# Generate normally distributed data
mean = 0
std_dev = 1
data = np.random.normal(mean, std_dev, 1000)

# Calculate variance
variance = np.var(data)
print(f"Variance: {variance}")

# Calculate entropy
from scipy.stats import norm

# Fit a normal distribution to the data
dist = norm(mean, std_dev)
entropy_normal = dist.entropy()
print(f"Entropy: {entropy_normal}")

# Plot density
plt.figure(figsize=(10, 6))
sns.histplot(data, kde=True, stat="density", linewidth=0)
plt.title('Density Plot of Normally Distributed Data')
plt.xlabel('Value')
plt.ylabel('Density')
plt.grid(True)
plt.show()
```

## Mixture of Gaussians

```python
from scipy.stats import norm
import seaborn as sns

# Generate mixture of normally distributed data
mean1, std_dev1, weight1 = 10, 5, 0.5
mean2, std_dev2, weight2 = 40, 5, 0.5

data1 = np.random.normal(mean1, std_dev1, int(1000 * weight1))
data2 = np.random.normal(mean2, std_dev2, int(1000 * weight2))
data_mixture = np.concatenate([data1, data2])

# Calculate variance
variance_mixture = np.var(data_mixture)
print(f"Variance of mixture: {variance_mixture}")

# Calculate entropy

# Fit normal distributions to the data
dist1 = norm(mean1, std_dev1)
dist2 = norm(mean2, std_dev2)

# Calculate the entropy of the mixture
entropy_mixture = weight1 * dist1.entropy() + weight2 * dist2.entropy()
print(f"Entropy of mixture: {entropy_mixture}")

# Plot density
plt.figure(figsize=(10, 6))
sns.histplot(data_mixture, kde=True, stat="density", linewidth=0)
plt.title('Density Plot of Mixture of Normally Distributed Data')
plt.xlabel('Value')
plt.ylabel('Density')
plt.grid(True)
plt.show()
```
