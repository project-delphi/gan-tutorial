# Unnormalized Probability

> Don't need normalized probability to simulate a random variable.

Probabiltiies can be defined using a probability density function (pdf). The pdf is a function that describes the likelihood of a random variable taking on a particular value. The pdf is defined such that the area under the curve is equal to 1. The probability of a random variable taking on a value within a range is equal to the area under the curve within that range.

Some authors will refer to the discrete case as a probability mass function (pmf) and the continuous case as a pdf. However, in this notebook, I will use the term pdf to refer to both the discrete and continuous cases.

# What is a bona fide pdf?

A bona fide pdf must satisfy the following properties:

1. The pdf must be non-negative for all values of the random variable.
2. The area under the curve must be equal to 1.
3. The probability of a random variable taking on a value within a range is equal to the area under the curve within that range.

In Latex, the pdf is denoted as $f(x)$, where $x$ is the random variable.

A bona fide pdf must satisfy the following properties:

$$f(x) \geq 0$$
$$\int_{x \in \text{Support}} f(x) dx = 1$$


A random variable $X$ that follows $f$ (denoted by $X \sim f$) is said to be distributed according to the pdf $f$. As a mapping X: $\Omega \rightarrow \text{Support}$, the pdf $f$ is a function that maps the sample space $\Omega$ to a support space of $X$, Support, usually a subset of $\mathbb{R}^k$.

## Some Example Probability Density Functions

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

def plot_binomial_pdf(n, p):
    x = np.arange(0, n+1)
    pmf = binom.pmf(x, n, p)
    
    plt.bar(x, pmf, color='blue')
    plt.xlabel('Number of successes')
    plt.ylabel('Probability')
    plt.title(f'Binomial Distribution (n={n}, p={p})')
    plt.show()

# Example usage
plot_binomial_pdf(10, 0.5)
```

```python
# Verify that pmf sums to close to 1 for a few small cases
def verify_binomial_pmf(n, p):
    x = np.arange(0, n+1)
    pmf = binom.pmf(x, n, p)
    pmf_sum = np.sum(pmf)
    print(f'Sum of PMF for n={n}, p={p}: {pmf_sum}')

# Example usage
verify_binomial_pmf(10, 0.5)
verify_binomial_pmf(5, 0.2)
verify_binomial_pmf(7, 0.8)
```

## Unnormalized pdf

Asking for a normalized pdf can be too much in some cases. In fact, just proportions can be enough to make decisions. For example, in the case of a Gaussian distribution, the pdf is given by:

$$f(x) = \frac{1}{\sqrt{2\pi\sigma^2}}e^{-\frac{(x-\mu)^2}{2\sigma^2}}$$

The term $\frac{1}{\sqrt{2\pi\sigma^2}}$ is a normalization constant that ensures that the area under the curve is equal to 1. However, if we are only interested in the relative probabilities of different values of $x$, we can ignore the normalization constant. This is called an unnormalized pdf.


Here's a simple example for cancer odds ratio. The odds ratio is a measure of association between an exposure and an outcome. The odds ratio is defined as the ratio of the odds of an event occurring in one group to the odds of it occurring in another group. The odds ratio is a common measure of association in epidemiology.


### Diabetes Odds Ratio

Example: Diabetes Odds Ratio

Suppose we are studying the association between diabetes and obesity. We have the following data:

- The odds of having diabetes if you are obese is 0.5. As a ratio this would be 1:2. As a probability distribution this is, in a table:

| Diabetes | Not Diabetes |
|----------|--------------|
| 1/3      | 2/3          |

Here the normalization constant is 3.

- The odds of having diabetes if you are not obese is 0.1. The odds ratio is 1:9. As a probability distribution this is, in a table:

| Diabetes | Not Diabetes |
|----------|--------------|
| 1/10     | 9/10         |

Here the normalization constant is 10.

Ratio is enough to make decisions. We can ignore the normalization constant. This is called an unnormalized pdf.

A normalzied probability mass function (pmf) sums to 1. It can be rewritten as unnormalized probability mass function (pmf) by multiplying with a constant. The constant is the reciprocal of the sum of the pmf. Take the example of the normal distribution. The normal distribution is given by:


$$f(x) = \frac{1}{\sqrt{2\pi\sigma^2}}e^{-\frac{(x-\mu)^2}{2\sigma^2}}$$


The term $\frac{1}{\sqrt{2\pi\sigma^2}}$ is a normalization constant that ensures that the area under the curve is equal to 1. However, if we are only interested in the relative probabilities of different values of $x$, we can ignore the normalization constant. This is called an unnormalized pdf.



```python
import numpy as np
from scipy.stats import norm

import matplotlib.pyplot as plt

# Parameters for the normal distribution
mu = 0  # mean
sigma = 1  # standard deviation

# Define the range of x values
x = np.linspace(-5, 5, 1000)

# Normalized PDF
normalized_pdf = norm.pdf(x, mu, sigma)

# Unnormalized PDF (ignoring the normalization constant)
unnormalized_pdf = np.exp(-0.5 * ((x - mu) / sigma) ** 2)

# Plotting the PDFs
plt.plot(x, normalized_pdf, label='Normalized PDF')
plt.plot(x, unnormalized_pdf, label='Unnormalized PDF', linestyle='--')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.title('Normalized and Unnormalized PDFs for Normal Distribution')
plt.legend()
plt.show()
```
