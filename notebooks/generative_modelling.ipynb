```python
from sklearn.datasets import make_moons
from sklearn.mixture import GaussianMixture
import matplotlib.pyplot as plt

# Generate moon-shaped data
X_moons, y_moons = make_moons(n_samples=300, noise=0.1, random_state=42)

# Fit a Gaussian Mixture Model
gmm = GaussianMixture(n_components=6, random_state=42)
gmm.fit(X_moons)

# Generate 10 new examples
new_samples, _ = gmm.sample(30)

# Plot the original data and the new samples
plt.scatter(X_moons[:, 0], X_moons[:, 1], c=y_moons, cmap='viridis', marker='o', label='Original Data')
plt.scatter(new_samples[:, 0], new_samples[:, 1], c='red', marker='x', label='Generated Samples')

# Add labels and title
plt.xlabel('x_0')
plt.ylabel('x_1')
plt.title('Moon-shaped Data and GMM Generated Samples')

# Show legend
plt.legend()

# Display plot
plt.show()
```
